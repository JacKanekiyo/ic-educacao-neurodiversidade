```python
print(f"Iniciando filtragem final do {CAMINHO_ENTRADA}...")
```

```python
try:
    # 2. Leitura do arquivo JÁ LIMPO
    # Não precisamos de tratamento de NaNs ou tipos, pois isso já foi feito.
    df = pd.read_csv(CAMINHO_ENTRADA, sep=',')

    # 3. Seleção e Filtro das Colunas
    # Usa df.filter para manter APENAS as colunas que estão na nossa lista padrão.
    df_filtrado = df.filter(items=COLUNAS_MANTER_FINAL)

    colunas_removidas = df.shape[1] - df_filtrado.shape[1]

    print(f"Colunas filtradas com sucesso. Foram removidas {colunas_removidas} colunas (ruído).")
    print(f"Base final com {df_filtrado.shape[0]} registros e {df_filtrado.shape[1]} colunas.")

    # 4. Salvar o arquivo final e padronizado
    df_filtrado.to_csv(CAMINHO_SAIDA, index=False)
    print(f"\n SUCESSO TOTAL: Ensino Médio padronizado e salvo em: {CAMINHO_SAIDA}")

except FileNotFoundError:
    print(f" ERRO: Verifique se o arquivo '{CAMINHO_ENTRADA}' está no local especificado.")
except Exception as e:
    print(f"\n ERRO FATAL: {e}")
```
